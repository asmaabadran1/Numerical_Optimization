#### Use your code from practical session 1 and create a function to perform GD for single variable linear regression.
#### The function should take the following input parameters:
##### Input data (X), Target Labels (y), Learning Rate (alpha), Maximum number of iterations
#### The function should return the following:
##### All Theta_0 and Theta_1 values till converge, and equivellant Losses, all hypothesis output, and optimum values of Theta_0 and Theta_1
#### Implement the stop criteria (Gradient Check).
#### Implement the convergence check (cost check).
#### You can use any small values for stop criteria and convergence check. However, the used values in the current implementation used 0.001 for both.

```python
def GDB(X,y,alpha,max_iteration):

    count = 0
    theta_0 = 0
    theta_1 = 0
    m = len(X)
    loss = []
    All_Theta_0 = []
    All_Theta_1 = []

    while count < max_iteration:




        print(f' ****************** Iteration {count} ********************')

        y_pred = theta_0 + theta_1 *  X
        print("h(x) :  ")
        print(y_pred)

        Error_Vector = (y_pred - y)
        print("Error_Vector :   ")
        print(Error_Vector)
        cost =  ( Error_Vector.dot(Error_Vector)) /2/m
        print("j =",cost)


        loss.append(cost)
        d_theta_0 = (1/m) * sum((y_pred - y))
        d_theta_1 = (1/m) * np.dot(y_pred - y , X)
        Gradient_Vector = np.array([d_theta_0,d_theta_1]).reshape(-1, 1)
        print("Gradient Vector:"  )
        print(Gradient_Vector)

        norm = np.linalg.norm([d_theta_0,d_theta_1])
        print("Gradient Vector Norm:   ",norm)


        theta_0_new = theta_0 - alpha * d_theta_0
        theta_1_new = theta_1 - alpha * d_theta_1
        theta_0 = theta_0_new
        print("theta_0_new :     ")
        print(theta_0)
        theta_1 = theta_1_new
        print("theta_0_new :     ")
        print(theta_1)
        All_Theta_0.append(theta_0)
        All_Theta_1.append(theta_1)
        count += 1
        if norm < 0.001:
                break

        if count>1:
            # Stopping using cost
            if abs(loss[-1]-loss[-2]) < 0.001:
              break



    return theta_0 , theta_1 ,loss , count , All_Theta_0,All_Theta_1,Error_Vector,y_pred,cost




theta_0 , theta_1 ,loss , count , All_Theta_0,All_Theta_1,Error_Vector,y_pred,cost= GDB(X,y,0.0005,300)
print("****************** Training Report ********************")

print(f'Gradient Descent converged after {count} epochs')
print("theta_0_Opt",theta_0)
print("theta_1_Opt",theta_1)

print("Error Vector:",Error_Vector)
print("cost",cost)
print("h(x) = y_predict:",y_pred)
print("y_actual:",y)
```

## Stochastic GD

#### Update the implementation of the Batch GD to be Stochastic and rpeat the above plots and evaluations
#### You can try different learning rate and see its effect.

##### Very Important notes:
- <font color='red'>Save your losses and thetas for each iterations (not each epoch).</font>
- <font color='red'>Plot loss vs. iterations (not epochs). This will allow you to see the difference between SGD (stochastic GD) and batch GD.</font>
- <font color='red'>Convergence check should be performed each epoch not each iterations. This allows you to perform more accurate decision. This is because the noisy process could lead to a deciving decision.</font>
- <font color='red'>Your result may be different from the results shown below due to the randomness of the process. As well as, the random shuffle. The results will be different also for two different run of the algorithm.</font>


##### The following results uses alpha = 0.0005 and max. iterations = 300.

```python
def SGD(X,y,alpha,epoch):
    theta_0=0
    theta_1=0
    m=len(X)
    loss_St=[]
    all_theta_0_St=[]
    all_theta_1_St=[]
    allpredictions=[]
    cost_per_epoch=[]

    shuffled_indices = np.random.permutation(len(X))
    X = X[shuffled_indices]
    y = y[shuffled_indices]

    for i in range(epoch):
        print(f' ****************** Epoch {i} ********************')
        for j in range(m):
            y_predictd=theta_0+X[j]*theta_1
            cost = 1/2 * (y_predictd-y[j])**2

            d_theta_0=y_predictd-y[j]
            d_theta_1=(y_predictd-y[j])*X[j]

            Gradient_Vector = np.array([d_theta_0,d_theta_1]).reshape(-1, 1)
            norm = np.linalg.norm([d_theta_0,d_theta_1])

            theta_0 =theta_0-(alpha*d_theta_0)
            theta_1 =theta_1-(alpha*d_theta_1)

            loss_St.append(cost)

            all_theta_0_St.append(theta_0)
            all_theta_1_St.append(theta_1)

            y_predictd=theta_0+X*theta_1
            allpredictions.append(y_predictd)

        print("Cost =",cost)
        print()
        print("Gradient Vector:")
        print(Gradient_Vector)
        print()
        print("Gradient Vector Norm:")
        print(norm)
        print()
        print("theta_0_new : ",theta_0)
        print("theta_1_new : ",theta_1)
        print()
        print("r2_score :", r2_score(y,y_predictd))
        cost_per_epoch.append(loss_St[-1])
        if i>=1:
            if np.linalg.norm(Gradient_Vector)<0.001 or abs(cost_per_epoch[-1]-cost_per_epoch[-2])<.001:
                break



    return cost_per_epoch,i,theta_0,theta_1,y_predictd,all_theta_0_St,all_theta_1_St,loss_St,cost

```

```python
cost_per_epoch,i,theta_0,theta_1,y_predictd,all_theta_0_St,all_theta_1_St,loss_St,cost=SGD(X,y,.0005,300)
```

```python
def plot_loss_3(losses, style):
    plt.figure(figsize=(10, 5))
    itration=list(range(len(losses)))
    plt.plot(itration,losses,color="mediumorchid")
    plt.xlabel("iteration")
    plt.ylabel("LOSS")
    plt.grid()

def plot_theta_loss(theta_vals, theta_name, losses, title, style):
    plt.figure(figsize=(10, 5))
    plt.plot(theta_vals, losses, style)
    plt.title(title)
    plt.xlabel(theta_name)
    plt.ylabel('Loss')
    plt.grid()

def plot_best_fit(theta0, theta1):
    plt.figure(figsize=(10, 5))
    plt.scatter(X, y)
    plt.plot(X, theta0 + theta1 * X, '-r')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.grid()
```

```python
plot_loss_3(loss_St,'-mo')
plot_theta_loss(all_theta_0_St,'Theta 0',loss_St,'Learning Rate = 0.0005','-bo')
plot_theta_loss(all_theta_1_St,'Theta 1',loss_St,'Learning Rate = 0.0005','-bo')
plot_best_fit(theta_0,theta_1)
```

## Mini-Batch GD

#### Update the implementation of the Batch GD to be Mini-Batch and repeat the above plots and evaluations
#### Change batch size and see the effect
#### Do not forget to shuffle the data

##### Very Important notes:
- <font color='red'>Save your losses and thetas for each iterations (not each epoch).</font>
- <font color='red'>Plot loss vs. iterations (not epochs). This will allow you to see the difference between SGD (stochastic GD) and batch GD.</font>
- <font color='red'>Convergence check should be performed each epoch not each iterations. This allows you to perform more accurate decision. This is because the noisy process could lead to a deciving decision.</font>
-<font color='red'>Your result may be different from the results shown below due to the randomness of the process. As well as, the random shuffle. The results will be different also for two different run of the algorithm.</font>

##### The following results uses batch size=5, alpha = 0.0005 and max. iterations = 300.

```python
def MB(X,y,alpha,num_min_batch,epoch):
    theta_0=0
    theta_1=0
    m=len(X)
    loss_St=[]
    all_theta_0_St=[]
    all_theta_1_St=[]
    allpredictions=[]
    cost_per_epoch=[]

    shuffled_indices = np.random.permutation(len(X))
    X = X[shuffled_indices]
    y = y[shuffled_indices]

    for i in range(epoch):
        print(f' ****************** Epoch {i} ********************')
        for j in range(0,m,num_min_batch):

            y_predictd=theta_0+X[j:j+num_min_batch].dot(theta_1)
            cost = np.sum(np.square(y_predictd-y[j:j+num_min_batch]))/(2*len(X[j:j+num_min_batch]))


            d_theta_0 = (1 / len(X[j:j+num_min_batch])) * np.sum(y_predictd - y[j:j+num_min_batch])
            d_theta_1 = (1 / len(X[j:j+num_min_batch])) * np.sum((y_predictd - y[j:j+num_min_batch]) @ X[j:j+num_min_batch])

            Gradient_Vector = np.array([d_theta_0,d_theta_1]).reshape(-1, 1)
            norm = np.linalg.norm([d_theta_0,d_theta_1])

            theta_0 =theta_0-(alpha*d_theta_0)
            theta_1 =theta_1-(alpha*d_theta_1)

            loss_St.append(cost )
            all_theta_0_St.append(theta_0)
            all_theta_1_St.append(theta_1)
            allpredictions.append(y_predictd)
        cost_per_epoch.append(cost)
        print("Cost =",cost)
        print()
        print("Gradient Vector:")
        print(Gradient_Vector)
        print()
        print("Gradient Vector Norm:")
        print(norm)
        print()
        print("theta_0_new : ",theta_0)
        print("theta_1_new : ",theta_1)
        print()

        if i>=1:
            if np.linalg.norm(Gradient_Vector)<0.001 or abs(cost_per_epoch[-1]-cost_per_epoch[-2])<0.001:
                y_predictd=theta_0+X.dot(theta_1)
                break
    print("r2_score :", r2_score(y,y_predictd))
    return cost_per_epoch,i,theta_0,theta_1,y_predictd,all_theta_0_St,all_theta_1_St,loss_St,cost
```

```python
cost_per_epoch,i,theta_0,theta_1,y_predictd,all_theta_0_St,all_theta_1_St,loss_St,cost=MB(X,y,0.0005,5,300)
```

```python
plot_loss_3(loss_St,'-mo')
plot_theta_loss(all_theta_0_St,'Theta 0',loss_St,'Learning Rate = 0.0005','-bo')
plot_theta_loss(all_theta_1_St,'Theta 1',loss_St,'Learning Rate = 0.0005','-bo')
plot_best_fit(theta_0,theta_1)
```
